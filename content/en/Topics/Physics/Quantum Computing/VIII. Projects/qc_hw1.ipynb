```python
import qiskit
from qiskit_experiments.framework import ParallelExperiment
from qiskit_experiments.library import StateTomography
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.transpiler import PassManager
from qiskit.visualization import plot_state_city

# GHZ State preparation circuit
nq = 2
qc_ghz = qiskit.QuantumCircuit(nq)
qc_ghz.h(0)
for i in range(1, nq):
    qc_ghz.cx(0, i)

# Draw the circuit
qc_ghz.draw(output='mpl')
```

```python

service = QiskitRuntimeService()
backend = service.least_busy(
    operational=True, min_num_qubits=qc_ghz.num_qubits, simulator=False
)
print(backend)

# QST Experiment
qstexp1 = StateTomography(qc_ghz)
qstdata1 = qstexp1.run(backend, seed_simulation = 100).block_for_results()

# Print total results
for result in qstdata1.analysis_results():
    print(result)

# Visualize the density matrix above
plot_state_city(qstdata1.analysis_results("state").value, title='Density Matrix')
```

```python
from qiskit import quantum_info
import numpy as np

# Define your data arrays
data_a = [1/np.sqrt(3), 1/np.sqrt(3), 0, 1/np.sqrt(3)]
data_b = [2/np.sqrt(3), 1/np.sqrt(3), -1/np.sqrt(3), -2/np.sqrt(3)]
data_c = [
    2/np.sqrt(13),  1/np.sqrt(13), -1/np.sqrt(13),
    -2/np.sqrt(13), 1/np.sqrt(13),  0,
    1/np.sqrt(13),  0,             np.sqrt(13)
]

def pretty_print_matrix(title, obj):
    """
    Prints a matrix or Qiskit object (with a .data attribute) using numpy's array_str
    for consistent formatting.
    """
    print(title)
    # Use the .data attribute if available
    matrix = obj.data if hasattr(obj, 'data') else obj
    print(np.array_str(np.array(matrix), precision=3, suppress_small=True))
    print()

def pretty_print_schmidt(schmidt_result):
    """
    Pretty prints the result of the Schmidt decomposition.
    If the result is a tuple, each component is printed with a header.
    """
    print("Schmidt Decomposition:")
    if isinstance(schmidt_result, tuple):
        for idx, comp in enumerate(schmidt_result):
            print(f"Component {idx}:")
            # Use .data if available, otherwise convert to a numpy array
            comp_array = comp.data if hasattr(comp, 'data') else np.array(comp)
            print(np.array_str(comp_array, precision=3, suppress_small=True))
            print()
    else:
        print(schmidt_result)
    print()

def analyze_state(data, dims, subsystem_indices):
    """
    Creates a Statevector from the provided data and then computes and prints:
        - Partial trace,
        - Schmidt decomposition,
        - Density matrix.
    The output for each operation is formatted for readability.
    """
    state = quantum_info.Statevector(data, dims=dims)
    
    for idx in subsystem_indices:
        print(f"--- Analyzing subsystem index: {idx} ---")
        
        # Compute and print the partial trace
        partial_trace_result = quantum_info.partial_trace(state, qargs=[idx])
        pretty_print_matrix("Partial Trace:", partial_trace_result)
        
        # Compute and print the Schmidt decomposition
        schmidt_decomp_result = quantum_info.schmidt_decomposition(state, qargs=[idx])
        pretty_print_schmidt(schmidt_decomp_result)
        
        # Compute and print the density matrix
        density_matrix_result = quantum_info.DensityMatrix(data, dims=dims)
        pretty_print_matrix("Density Matrix:", density_matrix_result)
        
        print("-" * 50)
    print("\n")

# -------------------------
# Analyze data_a and data_b for a 2-qubit system (dims=None)
# -------------------------
analyze_state(data_a, dims=None, subsystem_indices=range(2))
analyze_state(data_b, dims=None, subsystem_indices=range(2))

# -------------------------
# Analyze data_c as a tripartite system (dims=[3, 3, 1]) so that subsystems 0, 1, and 2 are defined.
# -------------------------
analyze_state(data_c, dims=[3, 3, 1], subsystem_indices=range(3))
```
